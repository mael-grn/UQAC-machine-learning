```python
# obtenir les vrais indices des colones de variables
# au lieu de fournir direcetement : 3, 4, 5, 6
rooms_ix, bedrooms_ix, population_ix, household_ix = [
    list(habitation.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", 
                "population", "households")]
```

```python
import numpy as np

def add_extra_features(X, add_bedrooms_per_room=True):
    rooms_par_habitation = X[:, rooms_ix] / X[:, household_ix]
    population_par_habitation = X[:, population_ix] / X[:, household_ix]
    if add_bedrooms_per_room:
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
        return np.c_[X, rooms_par_habitation, population_par_habitation,
                     bedrooms_per_room]
    else:
        return np.c_[X, rooms_par_habitation, population_par_habitation]

```

```python
from sklearn.preprocessing import FunctionTransformer
add_attribs= FunctionTransformer(add_extra_features, validate=False,
                                 kw_args={"add_bedrooms_per_room": False})

habitation_extra_attribs = add_attribs.fit_transform(habitation.values)
```

```python
habitation_extra_attribs
```

```python
habitation_extra_attribs = pd.DataFrame(
    habitation_extra_attribs,
    columns=list(habitation.columns)+["rooms_par_habitation", 
                    "population_par_habitation"],index=habitation.index)

habitation_extra_attribs.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', FunctionTransformer(add_extra_features, 
                                          validate=False)),
    ('std_scaler', StandardScaler())
    ])


habitation_num_tr = num_pipeline.fit_transform(habitation_num)
```

```python
habitation_num_tr
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(habitation_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

habitation_preparee= full_pipeline.fit_transform(habitation)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(habitation_preparee, habitation_labels)
```

```python
from sklearn.metrics import mean_squared_error
habitation_predictions = lin_reg.predict(habitation_preparee)
lin_mse = mean_squared_error(habitation_labels, habitation_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse 
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(habitation_preparee, habitation_labels)
```

```python
#reglage du modele : recherche par quadrillage
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = [
    # essayer 12 (3√ó4) combinaison d'hyperparametres
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # Apres essayer 6 (2√ó3) combinaison avec bootstrap = False
    {'bootstrap': [False], 'n_estimators': [3, 10], 
         'max_features': [2, 3, 4]},
  ]
forest_reg = RandomForestRegressor(random_state=42)
# entrainer a travers 5 folds, 
#ce qui donne un total de (12+6)*5=90 roundes d'entrainement 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error', 
                            return_train_score=True)
grid_search.fit(habitation_preparee, habitation_labels)
        
```

```python
#meilleur hyper-parametre
grid_search.best_params_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
#Evaluer votre systeme sur le jeu de test : test_set non nettoy√© 
#en appliquant le pipeiine. D'ou l'int√©r√™t du pipieline !
final_model = grid_search.best_estimator_

X_test = test_set.drop("median_house_value", axis=1)
y_test = test_set["median_house_value"].copy()

X_test_preparee = full_pipeline.transform(X_test)
final_predictions = final_model.predict(X_test_preparee)

final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```
